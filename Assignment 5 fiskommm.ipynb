Kerja yang dilakukan oleh gaya gesek merupakan bentuk kerja yang tidak diharapkan karena energi yang dikeluarkan, biasanya dalam bentuk panas atau bunyi yang dilepas ke lingkungan, tidak dapat dimanfaatkan lagi oleh sistem sehingga energi sistem berkurang.

## Gerak Benda di Atas Lantai Mendatar Kasar

Sistem yang ditinjau adalah suatu benda yang bergerak di atas lantai mendatar kasar. Benda diberi kecepatan awal tertentu dan bergerak melambat sampai berhenti karena adanya gaya gesek kinetis antara benda dan lantai kasar.

### Parameter 

Beberapa parameter yang digunakan adalah seperti pada tabel berikut ini.

Tabel 1. Simbol beserta satuan dan artinya

Simbol | Satuan | Arti

$x$ | m | posisi saat $t$

## Persamaan 

Persamaan-persamaan yang akan digunakan adalah seperti dicantumkan pada bagian ini.

### Kinematika

Hubungan antara antara kecepatan $v$, kecepatan awal $v_0$
, percepatan $a$, dan waktu $t$ diberikan oleh

\begin{equation}\label{eqn1}\tag{1}
v = v_0 + at
\end{equation}

Posisi benda $x$ bergantung pada posisi awal $x_0$, kecepatan awal 
$v_0$, percepatan $a$, dan waktu $t$ melalui hubungan

\begin{equation}\label{eqn2}\tag{2}
x = x_0 + v_0t + \frac{1}{2}at^2
\end{equation}

Selain kedua persamaan sebelumnya, terdapat pula persamaan berikut

\begin{equation}\label{eqn3}\tag{3}
v^2 = v_0^2 + 2a(x-x_0)
\end{equation}

yang menghubungkan kecepatan $v$ dengan kecepatan awal $v_0$, percepatan $a$, dan jarak yang ditempuh $x-x_0$.

### Dinamika

Hukum Newton I menyatakan bahwa benda yang semula diam akan tetap diam dan yang semula bergerak dengan kecepatan tetap akan tetap bergerak dengan kecepatan tetap bila tidak ada gaya yang bekerja pada benda atau jumlah gaya-gaya yang bekerja sama dengan nol

\begin{equation}\label{eqn4}\tag{4}
\sum F = 0
\end{equation}

Bila ada gaya yang bekerja pada benda bermassa $m$ atau jumlah gaya-gaya tidak nol

\begin{equation}\label{eqn5}\tag{5}
\sum F = ma
\end{equation}

maka keadaan gerak benda akan berubah melalui percepatan $a$, dengan $m$ > 0 dan $a \neq 0$

### Usaha

Usaha oleh suatu gaya $F$ dengan posisi awal $x_0$ dan posisi akhir $x$ dapat diperoleh melalui

\begin{equation}\label{eqn6}\tag{6}
W = \int_{x_0}^{x} F dx
\end{equation}

atau dengan 

\begin{equation}\label{eqn7}\tag{7}
W = \Delta K
\end{equation}

dengan $K$ adalah energi kinetik. Persamaan \eqref{eqn7} akan memberikan gaya oleh semua gaya. Dengan demikian bila $F$ adalah satu-satunya gaya yang bekerja pada benda, maka persamaan ini akan menjadi Persamaan \eqref{eqn6}

## Metode Numerik

Integrasi suatu fungsi $f(x)$ berbentuk 

\begin{equation}\label{eqn8}\tag{8}
A = \int_{a}^{b} f(x) dx
\end{equation}

dapat didekati dengan 

\begin{equation}\label{eqn9}\tag{9}
A \approx \sum_{i=0}^{N} f [\frac{1}{2}(x_i+x_{i+1} )]\Delta x 
\end{equation}

yang dikenal sebagai metode persegi titik tengah, dimana 

\begin{equation}\label{eqn10}\tag{10}
\Delta x = \frac{b-a}{N}
\end{equation}

dengan $N$ adalah jumlah partisi. Variabel $x_i$ pada Persamaan \eqref{eqn9} diberikan oleh 

\begin{equation}\label{eqn11}\tag{11}
x_i = a + i\Delta x
\end{equation}

dengan $i = 0,...,N$.

### Penyelesaian

Penerapan 1, 2, 3, 4, dan 5 pada gambar 2 akan menghasilkan

\begin{equation}\label{eqn12}\tag{12}
f_k = \mu_kmg
\end{equation}

dan usahanya adalah 

\begin{equation}\label{eqn13}\tag{13}
W = \int_{x_0}^{x} f_k dx
  = \int_{x_0}^{x} \mu_kmg dx
  = mg\int_{x_0}^{x} \mu_k dx
\end{equation}

dengan koefisien gesek statisnya dapat merupakan fungsi dari posisi $\mu_k = \mu_k(x)$
